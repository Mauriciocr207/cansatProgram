```python
import pandas as pd
#con panda se puede utilizar operaciones para las series

naranja = pd.Series([4,3,21,32,43])

print(naranja)
```

```python
#nombre de la serie[index de la serie a comenzar: recorrido que haga]
#con esto se puede leer elementos especificos de la serie
naranja[2:4]
```

```python
#Es para a conversion de una biblioteca a una serie,
#la cual se puede utilizar para guardar en nombres
materias = pd.Series({'espanol':12,'fisica':42,'mat':67})
print(materias)
```

```python
#con una biblioteca puedo llamar a los datos que esten guardados en el nombre
#en este caso es 12 guardado en "espanol"
materias['espanol']

#para llamar ambos datos, se requiere que este en doble corchete
materias[['espanol','mat']]
```

```python
#con valores en una serie, se pueden aplicar distintas funciones para
#poder hacer diferentes operaciones
numeros = pd.Series([1,2,3,4,5,6,7,8,9])
numeros.sum()
numeros.max()
numeros.min()
#funcion de desviacion estandar
numeros.std()
#te da un resumen de la serie de sus caracteristicas
numeros.describe()
print(numeros)
```

```python
#Tambien se puede filtrar los valores para que arroje solo aquellos que cumplan cierta condicion
print(numeros[numeros > 4])

#esta funcion permite ordenar referente a la cantidad numerica en orden descendente
print(materias.sort_values()) 

#esta permite ordenar los valores alfabeticamente indep del valor que contengan
print(materias.sort_index(ascending=True))
```
